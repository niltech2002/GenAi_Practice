```python
def sum_of_digits(n): 
    total = 0 
    while n > 0: 
        total += n % 10 
        n //= 10 
    print(total) 
 
sum_of_digits(1234) 
```

```python
while True: 
    print("1.Add") 
    print("2.Subtract") 
    print("3.Multiply") 
    print("4.Divide") 
    print("5.Exit") 
    
    choice = int(input("Enter choice: ")) 
 
    if choice == 5: 
        break 
 
    if choice not in [1,2,3,4]: 
        print("Invalid choice") 
        continue 
 
    a = float(input("Enter first number: ")) 
    b = float(input("Enter second number: ")) 
 
    if choice == 1: 
        print(a+b) 
    elif choice == 2: 
        print(a-b) 
    elif choice == 3: 
        print(a*b) 
    elif choice == 4: 
        print(a/b) 
```

```python
squares = [x*x for x in range(1,11)] 
print(squares)
 
even = [x for x in range(1,21) if x % 2 == 0] 
print(even)
 
chars = [ch for ch in "Python"] 
print(chars)
```

```python
nums = [10,20,30,40,50] 
 
target = int(input("Enter number: ")) 
 
for n in nums: 
    if n == target: 
        print("Found!") 
        break 
else: 
    print("Not found") 
```
