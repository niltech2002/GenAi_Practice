```python
import requests
import json
import os
from dotenv import load_dotenv
```

```python
# ─── Load environment variables from .env file ────────────────────────────────
load_dotenv(override=True)  # override=True allows .env values to overwrite existing env vars
```

```python
print(os.path.exists(".env"))
```

```python
# Accept dynamic user input
 
user_prompt = input("Enter your question: ")
```

```python
# API endpoint
 # ─── Configuration ────────────────────────────────────────────────────────────
API_KEY = os.getenv("ANTHROPIC_API_KEY")
url = "https://llmgw-wp.tekstac.com/v1/chat/completions"
 
# Headers
 
headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {API_KEY}"
}
 
# Payload
 
payload = {
    "model": "global.anthropic.claude-haiku-4-5-20251001-v1:0",
    "messages": [
        {
            "role": "user",
            "content": user_prompt
        }
    ],
    "temperature": 0.7
}
```

```python
# Send API request
 
response = requests.post(
    url,
    headers=headers,
    json=payload
)
 
print("Status Code:", response.status_code)
```

```python
# Print full JSON response
 
response_json = response.json()
 
print(json.dumps(response_json, indent=4))
```

```python
# Extract assistant response
 
assistant_reply = response_json["choices"][0]["message"]["content"]
 
print("\nAssistant Reply:")
print(assistant_reply)
```

```python
query1 = input("Enter Query 1: ")
query2 = input("Enter Query 2: ")
 
for query in [query1, query2]:
 
    payload["messages"] = [
        {
            "role": "user",
            "content": query
        }
    ]
 
    response = requests.post(
        url,
        headers=headers,
        json=payload
    )
 
    result = response.json()
 
    print("\nQuestion:", query)
    print("Answer:")
    print(result["choices"][0]["message"]["content"])
```
