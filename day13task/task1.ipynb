Hands-on Task 1: Extend the Basic Pipeline (No LLM) 

Objective 

Enhance an existing pipeline by adding one more processing step. 

Task 

Introduce a new node called word_count with the following functionality: 

Count the number of words in the input text 

Store the result in a new state field: 

word_count: int 

Update the pipeline flow as follows: 

START → count → word_count → summarise → END 

Modify the make_summary node so that the final output includes both: 

Character count 

Word count 

Expected Output 

Summary: 'LangGraph makes stateful AI workflows easy!' — 43 characters, 6 words. 

Skills Tested 

Extending TypedDict state 

Adding and connecting new nodes 

Passing state across nodes 

Using add_edge 

inp: LangGraph makes stateful AI workflows easy!
 

output:


Summary:
LangGraph makes stateful AI workflows easy!
 
Characters: 43
Words: 6
 

```python
#Step 1: Create State

from langgraph.graph import StateGraph, START, END
from typing import TypedDict
 
class GraphState(TypedDict):
    text: str
    char_count: int
    word_count: int
    summary: str
```

```python
#Step 2: Character Count Node
 
def count_characters(state):
    state["char_count"] = len(state["text"])
    return state
 
```

```python
#input
{
"text":"Hello World"
}
 
```

```python
#output
{
"text":"Hello World",
"char_count":11
}
```

```python
#Step 3: Word Count Node
 
def word_count(state):
    state["word_count"] = len(state["text"].split())
    return state
 
```

```python
{
"text":"Hello World",
"char_count":11,
"word_count":2
}
```

```python
#Step 4: Summary Node
 
def make_summary(state):
    state["summary"] = (
        f"{state['text']}\n"
        f"Characters: {state['char_count']}\n"
        f"Words: {state['word_count']}"
    )
    return state
 
```

```python
builder = StateGraph(GraphState)
```

```python
#Step 5: Build Graph
 
builder.add_node("count",count_characters)
builder.add_node("word_count",word_count)
builder.add_node("summary",make_summary)
 
```

```python
#Connect them

builder.add_edge(START,"count")
builder.add_edge("count","word_count")
builder.add_edge("word_count","summary")
builder.add_edge("summary",END)
```

#pipeline:

START
↓
count
↓
word_count
↓
summary
↓
END

```python
graph = builder.compile()
 
```

```python
result = graph.invoke({
"text":"LangGraph makes stateful AI workflows easy!"
})
 
```
