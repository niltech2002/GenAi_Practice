```python
import matplotlib.pyplot as plt
 
x = [1, 2, 3, 4, 5]
y = [5, 8, 6, 10, 7]
 
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
