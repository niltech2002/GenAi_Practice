```python
# ==========================================================
# Exercise 1: Text Embeddings & Similarity Comparison
# ==========================================================
 
# Import required libraries
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
# ----------------------------------------------------------
# Step 1: Define sample sentences
# ----------------------------------------------------------
sentences = [
    "GenAI is transforming software development",
    "Artificial Intelligence is changing how developers work",
    "I love playing cricket on weekends"
]
 
```

```python
# ----------------------------------------------------------
# Step 2: Load pre-trained Sentence Transformer model
# all-MiniLM-L6-v2 converts text into vector embeddings
# ----------------------------------------------------------
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# ----------------------------------------------------------
# Step 3: Generate embeddings
# Each sentence is converted into a numerical vector
# ----------------------------------------------------------
embeddings = model.encode(sentences)
```

```python
# ----------------------------------------------------------
# Step 4: Display first 5 dimensions of each embedding
# ----------------------------------------------------------
print("First 5 dimensions of each embedding:\n")
 
for i, embedding in enumerate(embeddings):
    print(f"Sentence {i+1}: {sentences[i]}")
    print(embedding[:5])  # Show only first 5 values
    print()
```

```python
# ----------------------------------------------------------
# Step 5: Calculate cosine similarity matrix
# ----------------------------------------------------------
similarity_matrix = cosine_similarity(embeddings)# ----------------------------------------------------------
# Step 5: Calculate cosine similarity matrix
# ----------------------------------------------------------
similarity_matrix = cosine_similarity(embeddings)
```

```python
# ----------------------------------------------------------
# Step 6: Display similarity matrix in readable format
# ----------------------------------------------------------
similarity_df = pd.DataFrame(
    similarity_matrix,
    index=[f"S{i+1}" for i in range(len(sentences))],
    columns=[f"S{i+1}" for i in range(len(sentences))]
)
 
print("Cosine Similarity Matrix:\n")
print(similarity_df)
```

```python
# ----------------------------------------------------------
# Step 7: Interpret Results
# ----------------------------------------------------------
print("\nInterpretation:")
print("Similarity close to 1 -> Sentences are semantically similar")
print("Similarity close to 0 -> Sentences are unrelated")
```
