```python

fruits = ["apple", "banana", "mango", "orange", "grape"]

fruits.append("pineapple")

fruits.remove("banana")

fruits[1] = "kiwi"

print("Final list:", fruits)
print("Length:", len(fruits))
```

```python

t = (1, 2, 3, 2, 4, 2)

print("Tuple:", t)

# Count occurrences
print("Count of 2:", t.count(2))

# Find index
print("Index of 3:", t.index(3))

#modifying (will give error)
try:
    t[0] = 10
except TypeError as e:
    print("Error:", e)
```

```python
# duplicates removed automatically
s = {1, 2, 3, 2, 4}
print("Set:", s)

# Add element
s.add(5)

# Remove using remove()
s.remove(3)

# Remove using discard()
s.discard(10)  # No error if not found

print("Updated set:", s)

# Try adding list (error)
try:
    s.add([6, 7])
except TypeError as e:
    print("Error:", e)
```

```python
# Create dictionary
d = {"name": "John", "email": "john@gmail.com", "phone": "1234567890"}

# Print keys and values
#print("Keys:", d.keys())
#print("Values:", d.values())
for key, value in d.items():
    print(key,":",value)

# Update name
d["name"] = "Mike"

# Add new key
d["address"] = "Pune"

# Access using get()
print("Email:", d.get("email"))

print("Updated dictionary:", d)
```

```python
# Create nested dictionary
student = {
    "name": "Alice",
    "marks": {"math": 90, "science": 85}
}

# Access nested value
print("Math marks:", student["marks"]["math"])

# Add new subject
student["marks"]["english"] = 88

# Print all
print("Student data:", student)
```

```python
veg = ["potato", "tomato", "onion", "carrot", "peas"]

a, b, *rest = veg

print("First:", a)
print("Second:", b)
print("Rest:", rest)
```

```python
# Membership
lst = [1, 2, 3, 4]
print(3 in lst)

# Identity
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  # True (values same)
print(a is b)  # False (different objects)
```

```python
import copy

# Nested list
original = [[1, 2], [3, 4]]

# Shallow copy
shallow = copy.copy(original)
shallow[0][0] = 99

# Deep copy
deep = copy.deepcopy(original)
deep[1][1] = 100

print("Original:", original)
print("Shallow:", shallow)
print("Deep:", deep)
```

```python
# Logical
print(True and False)
print(True or False)
print(not True)

# Bitwise
a = 5   # 101
b = 3   # 011

print(a & b)   # AND
print(a | b)   # OR
print(a ^ b)   # XOR
print(~a)      # NOT
print(a << 1)  # Left shift
print(a >> 1)  # Right shift
```

```python
# List loop
lst = [10, 20, 30]
for i in lst:
    print(i)

# Dictionary loop
d = {"a": 1, "b": 2}
for k in d:
    print(k)

for v in d.values():
    print(v)

for k, v in d.items():
    print(k, v)

# List of tuples
lst_tuples = [(1, 2), (3, 4)]
for x, y in lst_tuples:
    print(x, y)
```
