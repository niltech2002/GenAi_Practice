```python

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Data

x = np.arange(0, 100)
y = x * 2

# Create figure
fig = plt.figure()

# Add axes
ax1 = fig.add_axes([0, 0, 1, 1])        # Main axis
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2]) # Smaller inset axis

# Plot on both axes
ax1.plot(x, y)
ax2.plot(x, y)


# Labels for outer plot
ax1.set_xlabel('X Axis')
ax1.set_ylabel('Y Axis')
ax1.set_title('Outer Plot')

# Labels for inner plot
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Inner Plot')

# Show figure
plt.show()
```

#### Exercise 3
Create the plot below by adding two axes to a figure object at [0,0,1,1] and
[0.2,0.5,.4,.4]


```python

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Data
x = np.arange(0, 100)
y = x * 2
z = x ** 2

# Create figure
fig = plt.figure()

# Add axes
ax1 = fig.add_axes([0, 0, 1, 1])          # Main axis
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])  # Inset axis


# Main plot (x vs z)
ax1.plot(x, z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_title('Main Plot')

# Inset plot (x vs y)
ax2.plot(x, y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Zoomed Plot')

# Set limits for zoom effect
ax2.set_xlim(20, 40)
ax2.set_ylim(30, 120)

plt.show()
```
