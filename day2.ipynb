```python
test="default"  #global test
 
def check_scope():
    test="hello"  #non local test
    def do_local():
        test="local test"
 
    def do_nonlocal():
        nonlocal test
        test="non local test"
 
    def do_global():
        global test
        test="global test"
    do_local()
    print(test)
 
    do_nonlocal()
    print(test)
 
    do_global()
    print(test)
 
check_scope()
print(test)
```

```python
def function_to_be_used():
    print("This is  inside function")
 
def hello_decorator(func): #func-> function_to_be_used
    def inner():
        print("Before function execution")
        func()  
        print("After function execution")
    return inner
 
inn=hello_decorator(function_to_be_used)
inn()  #inn() -> inner()
```

```python
def hello_deco(func):
    def inner():
        print("**********")
        func()
        print("&&&&&&&&&&")
    return inner

@hello_deco
def our_home():
    print("this is our home")

our_home()
```

```python
def authentication_required(func):
    def wrapper():
        # Check if user is authenticated
        user_authenticated =  False  # Simulating an unauthenticated user
        if user_authenticated:            
            return func()
        else:            
            return "Access denied. Please log in to continue."
    return wrapper
 
@authentication_required
def my_secure_function():        
    return "Secret information only for authenticated users."
 
print(my_secure_function())
```
