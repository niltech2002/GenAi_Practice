```python
a = 1000
b = 1000
print(id(a))
print(id(b))
```

In Python, small integers (usually from -5 to 256) are interned. That means Python reuses the same object in memory for efficiency.
So a and b both point to the same object 10 in memory.
At this point:
id(a) == id(b)  # True
a is b          # True

```python
log_entry = '192.168.1.10 - - [09/Jun/2026:10:23:45 +0000] "GET /api/users HTTP/1.1" 200 1024'
parts = log_entry.split('"')
print(parts)
ip        = log_entry.split()[0]
request   = parts[1] # "GET /api/users HTTP/1.1"
method=request.split()[0]  # GET
endpoint  = request.split()[1]      # /api/users
status    = parts[2].strip().split()[0]     # 200

print(f"IP: {ip}")
print(request)
print(f"Method: {method}")
print(f"Endpoint: {endpoint}")
print(f"Status: {status}")
print(f"Is success: {status.startswith('2')}")
```

```python
# ── REAL-WORLD USE CASE: Employee Roster Management ──────────────────────────
employees = [
    {"id": 101, "name": "Alice",   "dept": "Engineering", "salary": 90000},
    {"id": 102, "name": "Bob",     "dept": "Marketing",   "salary": 72000},
    {"id": 103, "name": "Charlie", "dept": "Engineering", "salary": 95000},
    {"id": 104, "name": "Diana",   "dept": "HR",          "salary": 68000},
    {"id": 105, "name": "Eve",     "dept": "Engineering", "salary": 88000},
]

# Sort by salary descending
top_earners = sorted(employees, key=lambda e: e["salary"], reverse=True)
print("Top earners:")
for emp in top_earners[:3]:
   # print(emp['name'],"--->",emp['salary'])
   print(f"  {emp['name']:10} — ₹{emp['salary']:,}")

# Filter Engineering dept
eng_team = [e for e in employees if e["dept"] == "Engineering"]
print(eng_team)

avg_eng_salary = sum(e["salary"] for e in eng_team) / len(eng_team)
print(f"\nEngineering avg salary: ₹{avg_eng_salary:,.0f}")
```
